```python
import geopandas as gpd
import pandas as pd
```

```python
shp_kabkota = gpd.read_file('../SHP KABUPATEN INDONESIA/kab_20201_byindo.shp')
shp_kabkota['idkab'] = shp_kabkota['idkab'].astype(int)

def get_kota(kd):
    kd = str(kd)[0]
    if kd == '7':
        aa = 'KOTA'
    else:
        aa = ''

    return aa

shp_kabkota['klas'] = shp_kabkota['kdkab'].apply(get_kota)
shp_kabkota['nmkab'] = shp_kabkota['nmkab'] + ' ' + shp_kabkota['klas']


shp_kabkota = shp_kabkota[['idkab', 'nmprov', 'nmkab']]

shp_kabkota
```

```python
data = pd.read_csv('dataset simulasi.csv')
data = data[['idkab', 'adap_score', 'persen_anggaran_bencana', 'persen_anggaran_kesehatan', 'persen_anggaran_pangan', 'persen_anggaran_sampah','persen_anggaran_lh', 'persen_anggaran_jalan']]
data['idkab'] = data['idkab'].astype(int)
# data = data.dropna()

data
```

```python
data_merge = data.merge(shp_kabkota, on='idkab', how='left')
data_merge.to_csv('Simulasi.csv', sep=';', index=False)
data_merge
```
