```python
import geopandas as gpd
import pandas as pd
```

```python
shp_kabkota = gpd.read_file('kab_20201_byindo.shp')
shp_kabkota['idkab'] = shp_kabkota['idkab'].astype(int)

shp_kabkota
```

```python
shp_kabkota.columns
```

```python
shp_kabkota_df = shp_kabkota[['idkab', 'nmprov', 'nmkab', 'kdprov', 'kdkab']]
shp_kabkota_df = shp_kabkota_df[shp_kabkota_df['kdkab'] != '88'].reset_index(drop=True)
shp_kabkota_df
```

```python
shp_kabkota_df.to_csv('Data Kabkota BPS.csv', index = False)
```

```python
def get_kota(kd):
    kd = str(kd)[0]
    if kd == '7':
        aa = 'KOTA'
    else:
        aa = ''

    return aa

shp_kabkota['klas'] = shp_kabkota['kdkab'].apply(get_kota)
shp_kabkota['nmkab'] = shp_kabkota['nmkab'] + ' ' + shp_kabkota['klas']

shp_kabkota
```

```python
data_vul = pd.read_csv('vulnerability_index.csv')
data_vul = data_vul[['kab_code', 'expo_index', 'sensi_index', 'adap_index', 'vul_idx']]
data_vul['kab_code'] = data_vul['kab_code'].astype(int)

data_vul
```

```python
# MERGE SHP
desa_map_join = shp_kabkota.merge(data_vul, left_on='idkab', right_on='kab_code', how='left')
desa_map_join = desa_map_join[['idkab', 'nmprov', 'nmkab', 'kdprov', 'kdkab', 'geometry', 'expo_index', 'sensi_index', 'adap_index', 'vul_idx']]

desa_map_join.to_file("Merge/result.shp", index=False)

desa_map_join
```
